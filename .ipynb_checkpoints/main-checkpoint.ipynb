# Examine number of NaN values in data

```python
# examine nan values in categorical data
total_count = numeric_data.count() + numeric_data.isna().sum()
numeric_data.isna().sum() / total_count

# -or-

numeric_nan_pct = numeric_data.isna().mean()
# only display columns with a nan% above zero
print(numeric_nan_pct[numeric_nan_pct > 0])
```

```python
# examine nan values in categorical data
total_count = categorical_data.count() + categorical_data.isna().sum()
categorical_data.isna().sum() / total_count

# -or-

categorical_nan_pct = categorical_data.isna().mean()
# only display columns with a nan% above zero
print(categorical_nan_pct[categorical_nan_pct > 0])
```

# Examine Categorical Summary Stats

```python
categorical_stats = pd.DataFrame(columns = ['column', 'values', 'values_count_incna', 'values_count_nona', 
                                       'num_miss', 'pct_miss'])
tmp = pd.DataFrame()

for c in categorical_data.columns:
    tmp['column'] = [c]
    tmp['values'] = [categorical_data[c].unique()]
    tmp['values_count_incna'] = len(list(categorical_data[c].unique()))
    tmp['values_count_nona'] = int(categorical_data[c].nunique())
    tmp['num_miss'] = categorical_data[c].isnull().sum()
    tmp['pct_miss'] = (categorical_data[c].isnull().sum()/ len(categorical_data)).round(3)*100
    categorical_stats = categorical_stats.append(tmp)
    
categorical_stats
```

```python
# Let's do an Ascending sort on the Numboer of Distinct Categories for each categorical Variables
categorical_stats.sort_values(by = 'pct_miss', inplace = True, ascending = False)

# And set the index to Column Names
categorical_stats.set_index('column', inplace = True)
categorical_stats
```

# Impute Missing Data

### Numeric Data

```python
# make copy to avoid changing original data
# NOTE: copy full dataset including training and test data to increase imputation accuracy
imputed_data = numeric_data.copy()

# make new columns indicating what will be imputed
cols_with_missing = (col for col in imputed_data.columns 
                                 if imputed_data[col].isnull().any())

for col in cols_with_missing:
    imputed_data[col + '_was_missing'] = imputed_data[col].isnull()
    
# Imputation
my_imputer = SimpleImputer()
# translate dataframe to numpy matrix for fit_transofrm method
#imputed_data = imputed_data.values
numeric_imputed_data_copy = pd.DataFrame(my_imputer.fit_transform(imputed_data))
numeric_imputed_data_copy.columns = imputed_data.columns

numeric_imputed_data = numeric_imputed_data_copy
numeric_imputed_data.loc[:,["LotFrontage", "LotFrontage_was_missing", "MasVnrArea", "MasVnrArea_was_missing"]].head(15)
```

### Categorical Data

```python
# import custom imputing class that handles categorical features
from SeriesImputer import SeriesImputer

# make copy to avoid changing original data
# NOTE: copy full dataset including training and test data to increase imputation accuracy
imputed_data = categorical_data.copy()

# make new columns indicating what will be imputed
cols_with_missing = [col for col in imputed_data.columns 
                                 if imputed_data[col].isnull().any()]

# replace missing values with most frequently occuring value
for col in cols_with_missing:
    imputer = SeriesImputer()
    imputer.fit(imputed_data[col])
    imputed_data[col + '_imputed'] = imputer.transform(imputed_data[col])

categroical_imputed_data = imputed_data
categroical_imputed_data.loc[categroical_imputed_data["Electrical"].isna() == True]
```
